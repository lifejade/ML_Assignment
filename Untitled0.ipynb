```python

import pandas as pd
import numpy as np

train = pd.read_csv('data/2020.AI.bike-train.csv')
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
x_tr=train.drop('count',axis=1)
y_tr=train['count']

from sklearn.model_selection import train_test_split 
x_train, test, y_train, y_test = train_test_split(x_tr, y_tr, test_size=0.30, random_state=1)
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
x_train_imp=imp_mean.fit_transform(x_train)
test_imp=imp_mean.transform(test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor()
knr.fit(x_train_imp,y_train)
```

```python
pred=knr.predict(test_imp)
```

```python
import matplotlib.pyplot as plt
plt.plot(pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("MAE : %.2f, MSE : %.2f, R2 : %.2f "%((mean_absolute_error(y_test,pred)), mean_squared_error(y_test,pred), r2_score(y_test, pred)))
```
