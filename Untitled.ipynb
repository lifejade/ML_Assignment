```python
# visibility 시각화
# train data
data = train.copy()
data['hour_bef_visibility'] = np.power(data['hour_bef_visibility'], 3)
data['hour_bef_visibility'] /= np.max(data['hour_bef_visibility'])
fig, axes = plt.subplots(1, 2, figsize=(12,6))
sns.histplot(data['hour_bef_visibility'], ax=axes[0])
sns.regplot(data=data, x='hour_bef_visibility', y='count', ax=axes[1])
```

```python
def set_visibility(x):
    if x > 0.8:
        return 2
    elif x > 0.36:
        return 1
    else:
        return 0
```

```python
data['visibility'] = data['hour_bef_visibility'].map(set_visibility)
```

```python
data.groupby('visibility')['count'].mean()
```

```python
# hubidity 시각화
# train data
data = train.copy()
# data['hour_bef_humidity'] = (data['hour_bef_humidity'] - np.mean(data['hour_bef_humidity'])) / np.std(data['hour_bef_humidity'])
data['count'] = np.power(data['count'], 1/3)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,6))
sns.histplot(data['hour_bef_humidity'], ax=axes[0])
sns.regplot(data=data, x='hour_bef_humidity', y='count', ax=axes[1])
```

```python
# ozone 시각화
# train data
data = train.copy()
data['hour_bef_ozone'] = np.power(data['hour_bef_ozone'], 1/2)
data['count'] = np.power(data['count'], 1/3)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,6))
sns.histplot(data['hour_bef_ozone'], ax=axes[0])
sns.regplot(data=data, x='hour_bef_ozone', y='count', ax=axes[1])
```

```python
# 미세먼지 시각화
# train data
data = train.copy()
data['hour_bef_pm10'] = np.log1p(data['hour_bef_pm10'])
data['count'] = np.power(data['count'], 1/3)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,6))
sns.histplot(data['hour_bef_pm10'], ax=axes[0])
sns.regplot(data=data, x='hour_bef_pm10', y='count', ax=axes[1])
```
